```python
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv('Creditcard_data.csv')
data.head()
```

```python
data.shape
```

```python
# Checking data imbalance
data.groupby('Class').size()
```

```python
X_f = data.drop(columns=['Class'])
Y = data['Class']
```

```python
X_f = X_f.drop('Time',axis=1)
X_f.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_f)
X_df = pd.DataFrame(X, columns=X_f.columns)
```

```python
# Performing the SMOTE oversampling for given dataframe.
smote = SMOTE()
X_res,Y_res = smote.fit_resample(X_df,Y)
```

```python
df_res = pd.concat([X_res,Y_res], axis=1)
```

```python
df_res.groupby('Class').size()
```

```python
#Perform simple random sampling

e = 0.05
z = 1.96
p = 0.5
sample_size = int(np.ceil((z ** 2 * p * (1 - p)) / (e ** 2)))
random_sample = df_res.sample(n=sample_size)
print(f"Sample size: {sample_size}")
print(random_sample)
```

```python
#Perform stratified sampling

strata = data.groupby('Class')
stratum_sizes = strata.size()
s = stratum_sizes
e = 0.05
z = 1.96
p = 0.5
sample_sizes = {}
for stratum, size in stratum_sizes.items():
    sample_sizes[stratum] = int(np.ceil((z ** 2 * p * (1 - p)) / ((e) ** 2)))
    
sample_sizes

stratified_sample = pd.concat([df_res.sample(n=sample_sizes[stratum]) for stratum, stratum_df in strata])
print(f"Sample size: {sample_size}")
print(stratified_sample)
```

```python
#systematic sampling
interval = 2

# perform systematic sampling
systematic_sample = df_res.iloc[::interval]

print(systematic_sample)
```

```python
# cluster sampling
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(df_res)
cluster_assignments = kmeans.labels_

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
clustering_sample = df_res[cluster_series.isin(selected_clusters)]

print(clustering_sample)
```

```python
import random
#Multi stage sampling
# First stage: select 20 clusters
clusters = random.sample(list(df_res['Class'].unique()), 2)

# Second stage: select 50 individuals from each cluster
sample = pd.DataFrame()
for cluster in clusters:
    individuals = df_res[df_res['Class'] == cluster].sample(n=50)
    sample = pd.concat([sample, individuals])

print(sample)
```
